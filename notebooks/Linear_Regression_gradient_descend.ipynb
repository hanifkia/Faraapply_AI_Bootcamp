```python
import numpy as np
```

```python
x = np.array([20, 30, 40, 55, 60, 80, 100]) # speed
y = np.array([100, 70, 60, 50, 35, 20, 10]) # number of vehicles
```

```python
# normalization / Scaling

x = (x - np.min(x)) / (np.max(x) - np.min(x))
y = (y - np.min(y)) / (np.max(y) - np.min(y))
```

```python
x
```

```python
y
```

```python
# y = b0 + b1*x
```

```python
# initialization
b0 = 0
b1 = 0

n = len(x)

learning_rate = 0.01
epochs = 1000

losses = []
for epoch in range(epochs):
    
    y_pred = b0 + b1*x    
    
    # optimization 
    
    db0 = (-1/n) * np.sum(y - y_pred)
    db1 = (-1/n) * np.sum(x * (y - y_pred))
    
    # update
    b0 = b0 - learning_rate * db0
    b1 = b1 - learning_rate * db1
    
    y_pred = b0 + b1*x
    loss = np.mean((y - y_pred)**2)
    
    losses.append(loss)
    if epoch % 100 == 0:
        print(f"Epoch {epoch} : loss = {loss}")
print(f"Epoch {epoch} : loss = {loss}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(epochs), losses, label="LOSS (MSE)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
```

```python
from sklearn.datasets import load_breast_cancer
```
