```python
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
xData = data['data']
yData = data['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest =  train_test_split(xData, yData, test_size=0.2, shuffle=True, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=20)
model.fit(xTrain, yTrain)
```

```python
y_pred = model.predict(xTest)
```

```python
y_pred
```

```python
yTest
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(yTest, y_pred)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(xTrain, yTrain)
```

```python
y_pred = model.predict(xTest)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
print(f"ACC: {accuracy_score(yTest, y_pred)}")
print(f"PRC: {precision_score(yTest, y_pred)}")
print(f"RCL: {recall_score(yTest, y_pred)}")
print(f"F1: {f1_score(yTest, y_pred)}")
```

```python
print(classification_report(yTest, y_pred))
```
