```python
import numpy as np
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
xData = data['data']
yData = data['target']
```

```python
np.unique(yData, return_counts=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(xData, yData, stratify=yData, random_state=42)
```

```python
import xgboost as xgb
# pip install xgboost
```

```python
model = xgb.XGBClassifier()
model.fit(xTrain, yTrain)
y_pred = model.predict(xTest)
print(classification_report(yTest, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model = RandomForestClassifier()
model.fit(xTrain, yTrain)
y_pred = model.predict(xTest)
print(classification_report(yTest, y_pred))
```

```python
model = DecisionTreeClassifier()
model.fit(xTrain, yTrain)
y_pred = model.predict(xTest)
print(classification_report(yTest, y_pred))
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(xTrain, yTrain)
y_pred = model.predict(xTest)
print(classification_report(yTest, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {
    'kernel':('linear', 'rbf'),
    'C':[1, 10],
    'gamma': ["auto", "scale"]
}
```

```python
model = SVC()
model
```

```python
clf = GridSearchCV(model, parameters)
```

```python
clf.fit(xTrain, yTrain)
```

```python
best_params = clf.best_params_
```

```python
best_params
```

```python
model = SVC(**best_params)
```

```python
model
```

```python
model.fit(xTrain, yTrain)
y_pred = model.predict(xTest)
print(classification_report(yTest, y_pred))
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek
# pip install imblearn
```

```python
ov = SMOTE(random_state=42)
```

```python
xTrain_resample, yTrain_resample = ov.fit_resample(xTrain, yTrain)
```

```python
xTrain_resample.shape
```

```python
model = DecisionTreeClassifier()
model.fit(xTrain_resample, yTrain_resample)

y_pred = model.predict(xTest)
print(classification_report(yTest, y_pred))
```
