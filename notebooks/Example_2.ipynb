```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./diabetes.csv")
```

```python
df
```

```python
xData = df.iloc[:,:-1]
yData = df.iloc[:,-1]
```

```python
xData
```

```python
yData
```

```python
yData.value_counts()
```

```python
xData = xData.map(lambda x: x if x!=0 else np.nan)
```

```python
xData.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(xData, yData, stratify=yData, random_state=1000)
```

```python
yTrain.value_counts()
```

```python
yTest.value_counts()
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=5)
```

```python
imputer.fit(xTrain)
```

```python
xTrain = pd.DataFrame(imputer.transform(xTrain), columns=xTrain.columns)
xTest = pd.DataFrame(imputer.transform(xTest), columns=xTrain.columns)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, classification_report
```

```python
model = DecisionTreeClassifier()
model.fit(xTrain, yTrain)
y_pred = model.predict(xTest)
```

```python
accuracy_score(yTest, y_pred)
```

```python
f1_score(yTest, y_pred)
```

```python
precision_score(yTest, y_pred)
```

```python
recall_score(yTest, y_pred)
```

```python
print(classification_report(yTest, y_pred))
```
